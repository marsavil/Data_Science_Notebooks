#### Proyecto Desafio 1: Ventas Online

Analizar los resultados de un evento con los clientes de una empresa de venta online. Se recopiló un conjunto de datos que contiene los clientes que más gastaron en productos durante los 5 días de ventas, que es la duración del evento. Este análisis identificará al cliente con la mayor compra esta semana, quien recibirá un premio de la tienda, y posteriormente, puede ayudar a la empresa a crear nuevas estrategias para atraer más clientes.

```python
# Quitar 'R$' y reemplazar ',' por '.', luego convertir a float
import pandas as pd

# Suponiendo que ya cargaste el DataFrame
def limpiar_valores(valores):
    return [float(v.replace('R$', '').replace(',', '.').strip()) for v in valores]

datos_ventas['Valor da compra'] = datos_ventas['Valor da compra'].apply(limpiar_valores)
```

```python
datos_ventas
```

```python
# APLANAR EL DATA FRAME EN UNA LISTA DE DICCIONARIOS
registros = []

for i, row in datos_ventas.iterrows():
    clientes = row['Cliente']
    valores = row['Valor da compra']
    for cliente, valor in zip(clientes, valores):
        registros.append({'Cliente': cliente, 'Valor': valor})

registros
```

```python
# CREAR UN NUEVO DATA FRAME
df_compras = pd.DataFrame(registros)

# SUMAR LOS VALORES POR CLIENTE
gastos_totales = df_compras.groupby('Cliente')['Valor'].sum()

# OBTENER EL CLIENTE CON MAYOR GASTO
cliente_top = gastos_totales.idxmax()
monto_top = gastos_totales.max()

print(f'Cliente que más gastó: {cliente_top} con un total de R$ {monto_top:.2f}')
```

```python
gastos_totales
```

```python
import matplotlib.pyplot as plt

top_5 = gastos_totales.sort_values(ascending=False).head(5)

plt.figure(figsize=(10, 6))
bars = plt.bar(top_5.index, top_5.values, color='skyblue')

# Agregar los valores sobre cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 10, f'R$ {yval:.2f}', ha='center', va='bottom')

plt.title('Top 5 clientes que más gastaron durante el evento')
plt.ylabel('Valor total de compras (R$)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


#
```

#### Proyecto Desafío 2: Administración de Condominios

Analizar el retraso en el pago del alquiler en el condominio de algunos residentes. El data frame datos_locacion contiene información sobre el departamento de los inquilinos, el día acordado para el pago del alquiler, el día en que se realiza el pago del alquiler y el monto del alquiler.

```python
datos_locacion
```

```python
from datetime import datetime

def procesar_fila(fila):
    fechas_pactadas = [datetime.strptime(d, "%d/%m/%Y") for d in fila['datas_combinadas_pagamento']]
    fechas_pagadas = [datetime.strptime(d, "%d/%m/%Y") for d in fila['datas_de_pagamento']]
    valores = [float(v.replace('$', '').replace('reais', '').replace(',', '.').strip()) for v in fila['valor_aluguel']]

    # Calcular días de retraso
    retrasos = [(pago - pactado).days for pactado, pago in zip(fechas_pactadas, fechas_pagadas)]

    return pd.Series({
        'apartamento': fila['apartamento'],
        'dias_retraso': retrasos,
        'valores': valores
    })

# Aplicar a cada fila
df_retrasos = datos_locacion.apply(procesar_fila, axis=1)

df_retrasos
```

```python
# clientes con más días de retraso

# Explode para tener una fila por pago
df_detalle = df_retrasos.explode(['dias_retraso', 'valores'])

# Agregar total de días de retraso por apartamento
df_agrupado = df_detalle.groupby('apartamento')['dias_retraso'].sum().sort_values(ascending=False)

print(df_agrupado.head())
```

```python
import matplotlib.pyplot as plt

df_agrupado.head(5).plot(kind='barh', figsize=(10, 5), color='tomato')
plt.xlabel("Total días de retraso")
plt.title("Top 5 apartamentos con mayor retraso en pagos")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Debido a una inestabilidad en el sitio web de la empresa, tuvimos problemas con los nombres de los clientes durante el guardado. Esto resultó en una columna de nombres de clientes con una combinación de letras, mayúsculas y minúsculas, números y otros caracteres.

Sabiendo esto, manipula los textos de la columna Cliente para que el resultado sean los nombres de los clientes en letras minúsculas, con ausencia de caracteres especiales o números.

```python
datos_ventas.head()
```

```python
datos_ventas['Cliente'] = datos_ventas['Cliente'].str.lower()
datos_ventas
```

```python
import unicodedata
import re

def limpiar_con_regex(texto):
    # Paso 1: eliminar acentos
    texto_normalizado = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('utf-8')
    # Paso 2: eliminar caracteres que no sean letras ni espacios
    return re.sub(r'[^a-zA-Z\s]', ' ', texto_normalizado)

```

1. unicodedata.normalize('NFKD', texto)
Esto descompone los caracteres con tildes o signos diacríticos en sus componentes base.

Ejemplo: "á" se convierte en "á" (letra "a" + tilde separado)

2. .encode('ASCII', 'ignore')
Convierte el texto a ASCII, ignorando cualquier carácter que no sea parte del alfabeto ASCII.

Eso elimina los caracteres "no normales", como la tilde que quedó de "á".

3. .decode('utf-8')
Convierte el resultado binario (bytes) de nuevo a una cadena de texto.

4. re.sub(r'[^a-zA-Z\s]', ' ', texto_normalizado)
Esto reemplaza todos los caracteres que NO sean:

a-z o A-Z: letras

\s: espacios

por un espacio ' '.

```python
datos_ventas['Cliente'] = datos_ventas['Cliente'].apply(limpiar_con_regex)
datos_ventas
```
