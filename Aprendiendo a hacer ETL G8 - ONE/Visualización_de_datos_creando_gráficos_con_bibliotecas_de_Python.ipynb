La gerencia de la empresa necesita visualizar claramente las tendencias de ventas en cada tienda, para que puedan tomar decisiones estratégicas sobre inventarios y acciones de marketing. Para ello, debes crear cuatro subgráficos dispuestos en dos filas y dos columnas, donde cada subgráfico represente una tienda diferente. En este desafío, cada subgráfico debe presentar un gráfico de líneas que muestre la variación del número de ventas a lo largo de los meses del año.

¡Ahora es el momento de mostrar tus habilidades en análisis de datos y visualización! Para crear el DataFrame con el número de ventas de las tiendas y crear la figura, utiliza la siguiente información:

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
```

Consejo: Para facilitar la creación de los subgráficos, puedes definir la columna "tiendas" como índice del DataFrame y utilizar la propiedad loc de la biblioteca Pandas para trazar cada una de las tiendas.

¡No olvides agregar un título general a la figura, títulos a los subgráficos y etiquetas a los ejes! Además, asegúrate del tamaño de la figura y del espaciado entre los subgráficos.

```python
df = pd.DataFrame(ventas_2022, index=tiendas)
months = list(ventas_2022.keys())
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Ventas mensuales de las 4 tiendas durante 2022')


ax[0,0].plot(df.loc['A', months], color='red')
ax[0,0].set_title('Ventas tienda A')

ax[0,1].plot(df.loc['B', months], color='green')
ax[0,1].set_title('Ventas tienda B')

ax[1,0].plot(df.loc['C', months], color='purple')
ax[1,0].set_title('Ventas tienda C')

ax[1,1].plot(df.loc['D', months], color='orange')
ax[1,1].set_title('Ventas tienda D')

y_min = 0
y_max = 700

for x in ax.ravel():
  x.set_ylim(y_min, y_max)
  x.set_xlabel('Mes')
  x.set_ylabel('Ventas')
  x.grid()


plt.show()
```

```python
paises = ['Brasil', 'Argentina', 'Perú', 'Colombia']
max_4 = sudamerica.loc[paises]
```

```python
# Aplicar estilo y paleta de colores de Seaborn
sns.set(style='whitegrid')
sns.set_palette('Set2')  # Paleta de colores suaves


df_tendencia = max_4[years].transpose() # crear un data frame donde los indices sean los años
df_tendencia.index = df_tendencia.index.astype(int)  # Convertir índices (años) a enteros
df_tendencia.columns.name = None  # Quitar nombre del eje de columnas si lo tiene

# Crear gráfico
fig, ax = plt.subplots(figsize=(12, 6))

# Dibujar una línea por cada país
for pais in df_tendencia.columns:
    ax.plot(df_tendencia.index, df_tendencia[pais], label=pais, linewidth=2.5, marker='o')

# Título y etiquetas
ax.set_title('Tendencia de inmigración a Canadá desde Brasil, Argentina, Perú y Colombia (1980-2013)',
             fontsize=18, loc='left')
ax.set_xlabel('Año', fontsize=14)
ax.set_ylabel('Número de inmigrantes', fontsize=14)

# Personalización
ax.legend(title='País', fontsize=12, title_fontsize=13)
ax.tick_params(axis='both', labelsize=12)
ax.grid(True, linestyle='--', alpha=0.6)

# Mejora de distribución
plt.tight_layout()
plt.show()
```

Consejos:

* Utiliza el código proporcionado para Brasil como base y adáptalo para incluir los datos de Argentina.
* Para configurar las animaciones, puedes hacer un bucle for para recorrer el DataFrame datos_brasil y, para cada iteración, crear una nueva lista que contenga dos objetos del tipo go.Scatter, uno para cada país. Luego, cada lista puede ser utilizada para crear un objeto go.Frame, que se agrega a la lista de frames. Finalmente, la lista de frames puede ser asignada al objeto fig, que es la figura del gráfico que se animará.
```python
datos_br['Año'] = datos_br['Año'].astype(int)
```

```python

import plotly.graph_objs as go

# Diccionario con DataFrames por país
datos = {
    'Argentina': datos_arg,
    'Brasil': datos_br
}

# Crear figura
fig = go.Figure()

# Colores opcionales por país (se puede omitir o cambiar)
colores = {
    'Argentina': 'blue',
    'Brasil': 'green'
}

# Agregar la línea inicial (solo primer punto) para cada país
for pais, df in datos.items():
    fig.add_trace(
        go.Scatter(
            x=[df['Año'].iloc[0]],
            y=[df['Inmigrantes'].iloc[0]],
            mode='lines',
            name=pais,
            line=dict(width=4, color=colores.get(pais, None))
        )
    )

# Crear frames para la animación (una por año)
frames = []

for i in range(1, len(datos_arg)):  # Suponemos que todos los DataFrames tienen el mismo número de años
    frame_data = []

    for pais, df in datos.items():
        frame_data.append(
            go.Scatter(
                x=df['Año'].iloc[:i+1],
                y=df['Inmigrantes'].iloc[:i+1],
                mode='lines',
                name=pais,
                line=dict(width=4, color=colores.get(pais, None))
            )
        )

    frames.append(go.Frame(data=frame_data, name=str(i)))

# Asignar frames a la figura
fig.frames = frames

# Configurar layout
fig.update_layout(
    title=dict(
        text='<b>Inmigración de Argentinos y Brasileños hacia Canadá (1980-2013)</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Año</b>'),
    yaxis=dict(range=[0, 7000], autorange=False, title='<b>Número de Inmigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1000,
    height=500
)

# Mostrar figura
fig.show()
```
