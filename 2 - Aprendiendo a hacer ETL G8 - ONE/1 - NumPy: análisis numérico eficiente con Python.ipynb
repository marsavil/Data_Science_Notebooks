### Sin embargo este resultado no es el adecuado ya que se hace por ensayo y error al determinar la pendiente de la recta

Para una solución correcta trabajaremos con estas variables y fórmulas.

$$\alpha = pendiente$$
$$\beta = interseccion$$
$$n = observaciones$$
$$y = precio$$
$$x = fecha$$


La solución para la pendiente es:

$$\alpha = \frac{n \sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2} $$

Y para la intersección:
 $$\beta = \frac{\sum y - \alpha \sum x}{n} $$

Y obtenemos la ecuación de la recta:

$$y = \alpha x + \beta $$

```python
n = np.size(Moscu)
y = Moscu
x = cantidad_fechas
```

pendiente:

$$\alpha = \frac{n \sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2} $$

```python
a = (n*np.sum(x*y)-np.sum(x)*np.sum(y))/(n*np.sum(x**2)-np.sum(x)**2)
a
```

intersección:

 $$\beta = \frac{\sum y - \alpha \sum x}{n} $$

```python
b = np.mean(y)- a*np.mean(x)
b
```

$$y = \alpha x + \beta $$

```python
y = a * x + b
y
```

```python
np.linalg.norm(Moscu - y)
```

```python
plt.plot(cantidad_fechas, Moscu)
plt.plot(x,y)
```

Este gráfico, si bien similar al anterior, es el correcto

Si quiero obtener un precio estimado (no real) en un momento preciso

```python
plt.plot(cantidad_fechas, Moscu)
plt.plot(x,y)
plt.plot(25,a*25+b, '*r')
```
