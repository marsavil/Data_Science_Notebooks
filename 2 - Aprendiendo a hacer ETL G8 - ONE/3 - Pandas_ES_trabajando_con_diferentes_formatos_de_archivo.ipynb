# Leyendo Banco de Datos

## Creando un banco de datos local

```python
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
id = '1fT_GbEIDmeF9yinfd2HuDTNEZTQVP-Ey'
url = f'https://drive.google.com/uc?id={id}'
clientes = pd.read_csv(url)
clientes
```

```python
with engine.connect() as conn:
    conn.execute(text("DROP TABLE IF EXISTS clientes"))  #Este código es por si hay que volver a crear la tabla

clientes.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
inspector.get_table_names()
```

## Leyendo una consulta SQL

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renta ="Empleado"'
```

```python
empleados = pd.read_sql(sql = text(query), con = engine.connect())
```

```python
empleados
```

```python
with engine.connect() as conn:
    conn.execute(text("DROP TABLE IF EXISTS empleados"))  #Este código es por si hay que volver a crear la tabla
empleados.to_sql('empleados', engine, index=False)
```

```python
pd.read_sql_table('empleados', engine)
```

### Borrando registros

```python
from sqlalchemy.exc import SQLAlchemyError
```

```python
query = 'DELETE FROM clientes WHERE ID_Cliente=5008804'

try:
  r_set = engine.connect().execute(text(query))
except SQLAlchemyError as e:
  print(e)
else:
  print('Se borró el registro', r_set.rowcount)
```

### Actualizar registros

```python
query = 'UPDATE clientes SET Grado_estudio = "Nivel superior" WHERE ID_Cliente=5008810'

try:
  r_set = engine.connect().execute(text(query))
except SQLAlchemyError as e:
  print(e)
else:
  print('Se actualizó el registro', r_set.rowcount)
```
